```python
import pandas as pd
import numpy as np
```

```python
file_paths = [
'/content/101.csv',
'/content/102.csv',
'/content/103.csv',
'/content/104.csv',
'/content/105.csv',
'/content/107.csv',
'/content/108.csv',
'/content/112.csv',
'/content/113.csv',
'/content/114.csv',
'/content/115.csv',
'/content/116.csv',
'/content/117.csv',
'/content/118.csv',
'/content/119.csv',
'/content/120.csv',
'/content/121.csv',
'/content/122.csv',
'/content/123.csv',
'/content/125.csv',
'/content/126.csv',
'/content/127.csv',
'/content/128.csv',
'/content/129.csv',
'/content/130.csv',
'/content/134.csv',
'/content/135.csv',
'/content/136.csv',
'/content/137.csv',
'/content/801.csv',
'/content/802.csv'
]

dfs={}

for i, file_path in enumerate(file_paths, start=1):
    df = pd.read_csv(file_path)
    dfs[f'df{i}'] = df
```

```python
# Mapping file read
file_path2 = '/content/NewGen AR_Past Period Mapping.xlsx'
mapping = pd.read_excel(file_path2)
mapping
```

```python
# Merge all csv files
merged_df = pd.concat(dfs, ignore_index=True)
merged_df
```

```python
# GL Account column / extract only necessary data

merged_df['GL Account'] = merged_df['GL Account'].astype(str)
merged_df['GL Account'] = merged_df['GL Account'].str[:15]
```

```python
# "Amount" 컬럼 값이 NaN인 행의 인덱스를 가져와서 해당 행의 "Days" 값을 "Amount"로 대체
nan_indices = pd.isna(merged_df['Amount'])
merged_df.loc[nan_indices, 'Amount'] = merged_df.loc[nan_indices, 'Days']

# Amount 컬럼의 쉼표 제거하고 숫자로 변환
merged_df['Amount'] = merged_df['Amount'].astype(float)

# 조건에 따라 debit 또는 credit 컬럼에 저장
merged_df['Debit'] = merged_df['Amount'].apply(lambda x: x if x > 0 else 0)
merged_df['Credit'] = merged_df['Amount'].apply(lambda x: -x if x < 0 else 0)

merged_df
```

```python
merged_df.columns
```

```python
# check whether there is NaN value in Amount column

merged_df[merged_df['Amount'].isna()]
```

```python
# JE_new 컬럼 추가
merged_df['JE_new'] = merged_df['JE Name'].str[:11]

# 'GL Account Desc' 열에서 알파벳과 빈 칸을 제외한 모든 문자를 빈 문자열로 대체
merged_df['GL Account Desc'] = merged_df['GL Account Desc'].str.replace('\t', '')

# 조건에 해당하는 행을 필터링하고 'New GL Account' 컬럼에 값을 입력
condition1 = (merged_df['GL Account Desc'] == 'Provision For Bad Debts') & (merged_df['Debit'] > 0)
merged_df.loc[condition1, 'New GL Account'] = '102800-0000-000'
```

```python
# Provision For Bad Debts 컬럼 불러오기
merged_df[merged_df['GL Account Desc']=="Provision For Bad Debts"]
```

```python
# Past Period이고, mapping에 해당 GL 있는 경우 해당하는 행을 필터링하고 'New GL Account' 컬럼에 값을 입력
condition2 = (merged_df['JE_new'] == 'Past Period') & (merged_df['GL Account'].isin(mapping['GL Account']))
merged_df['New GL Account'] = merged_df['GL Account'].map(mapping.set_index('GL Account')['New GL Account'])

# 'JE_new'가 'Past Period'이고 'GL Account'가 'mapping'에 없는 경우 'New GL Account'에 'Need research' 입력
condition3 = (merged_df['JE_new'] == 'Past Period') & (~merged_df['GL Account'].isin(mapping['GL Account']))
merged_df.loc[condition3, 'New GL Account'] = 'Need research'

# 'JE_new'가 'Past Period'가 아닌 경우 'New GL Account'에 'GL Account' 값을 입력
condition4 = merged_df['JE_new'] != 'Past Period'
merged_df.loc[condition4, 'New GL Account'] = merged_df.loc[condition4, 'GL Account']
```

```python
# New GL Account에 NaN 값 있는지 확인

merged_df[merged_df['New GL Account'].isna()]
```

```python
# description 컬럼 추가
merged_df['description'] = '09/2023 NewGen AR File'  # This should be updated for every month

# GL Code 컬럼 추가
merged_df['GL Code'] = merged_df['Facility Code'].astype(str) + '-' + merged_df['New GL Account']

merged_df = merged_df[['Facility Code', 'Facility Name', 'Period Year', 'Period Month','JE Code', 'JE Name','GL Account', 'GL Account Desc', 'Amount', 'Days',
       'New GL Account', 'description', 'GL Code','Debit', 'Credit' ]]
merged_df.head()
```
