```python
import pandas as pd
```

```python
# csv file read
file_path = '/content/invoice_list.csv'
file_path2 = '/content/CoA.csv'
invoice = pd.read_csv(file_path, encoding ='latin1')
coa = pd.read_csv(file_path2, encoding ='latin1')
```

```python
# prompt: change the column name

invoice.rename(columns={'ï»¿GL#':'GL_account', "Orig. Master Name" : 'vendor_name' }, inplace=True)
invoice['GL_account'] = invoice['GL_account'].str[4:20]
invoice.head()
```

```python
coa.rename(columns={'ï»¿GL Account':'GL_account'}, inplace=True)
coa['GL_account'] = coa['GL_account'].str[4:20]
coa = coa[['GL_account', 'Description']]
coa.head()
```

```python
# prompt: make a table with vendor_name and GL account column only

new_table = invoice[['vendor_name', 'GL_account']]
new_table.head()
```

```python
# prompt: recommend the most frequent GL_account for vendor_name

most_frequent_gl_account = new_table.groupby('vendor_name')[['GL_account']].agg(pd.Series.mode)
most_frequent_gl_account = most_frequent_gl_account.explode('GL_account')
most_frequent_gl_account = most_frequent_gl_account.groupby('vendor_name').head(3)

# Reset index to make 'vendor_name' a column
most_frequent_gl_account.reset_index(inplace=True)

# Print the most frequent GL account for each vendor
most_frequent_gl_account.head(20)
```

```python
# prompt: from coa table find GL_account and left join the value to the new_table

df = pd.merge(most_frequent_gl_account, coa, how='left', on='GL_account')
df.head(20)
```

```python
# prompt: if Description value is NaN change the GL_account value to first 7 digits and '0000-000'

df['GL_account'] = df.apply(lambda x: x['GL_account'][:7] + '0000-000' if pd.isna(x['Description']) else x['GL_account'], axis=1)
df.head(20)
```

```python
# prompt: from coa table find GL_account and left join the value to the new_table

df = pd.merge(df, coa, how='left', on='GL_account')
Vendor_GL_matching = df[['vendor_name', 'GL_account', 'Description_y']]
Vendor_GL_matching.rename(columns={'Description_y':'Description'}, inplace=True)
Vendor_GL_matching.head(20)
```
