```python
import pandas as pd
```

```python
# Bank file read
file_path = '/content/CSVEXP_MAY_FCB.csv'
bank = pd.read_csv(file_path, encoding='latin1')

# Check list from GP read
file_path2 = '/content/Checkbook_MAY.csv'
check = pd.read_csv(file_path2, encoding='latin1')

# Bank Account list read
file_path3 = '/content/BankAccountnumber.xlsx'
bankaccount = pd.read_excel(file_path3)
bankaccount.head(10)
```

```python
bank1 = bank[['Post Date', 'Account Number', 'Client Account Name', 'Transaction Description', 'Amount','Bank Reference','Customer Reference','Transaction Detail']]
bank1
```

```python
# From Bank file extract Cleared column only

bank1 = bank[['Post Date', 'Account Number', 'Client Account Name', 'Transaction Description', 'BAI Code', 'Amount','Debit/Credit','Bank Reference','Customer Reference','Transaction Detail','Status' ]]
bank1 = bank1[bank1['Status']=='Cleared']
```

```python
# Bankaccount for every entity / Operating account

bankaccount_subset = bankaccount[['Bank Account Summary', 'Unnamed: 1']]

# Change the 1st raw as a column name
bankaccount_subset.columns = bankaccount_subset.iloc[1]
bankaccount_subset = bankaccount_subset[3:].reset_index(drop=True)
bankaccount_subset.drop_duplicates(subset='Entity code', keep='first', inplace=True)

# Change Entity code to int type
bankaccount_subset['Entity code'] = bankaccount_subset['Entity code'].astype(int)
bankaccount_subset.head()
```

```python
# Check account number From entity 108 to 802
bankaccount_subset[
    (bankaccount_subset['Entity code'] >= 101) &
    (bankaccount_subset['Entity code'] <= 802)
]
```

```python
# Enter the account number you want to retrieve
accnt_numb = ['2337','2361','1975','2003','2046','2599','2644','1801','1879','1836','1895','1852','2417','2468','0796','2505','2556','2126','2142','2687','2177','2206','2230','2265','2302','0833','1438','1454','1542','2070','2724']

newbank = bank1[bank1['Account Number'].astype(str).str[-4:].isin(accnt_numb)]
newbank

# Sort necessary information according to account #, Description, Post Date
sorted_bank = newbank.sort_values(by=['Account Number', 'Transaction Description', 'Post Date'])
sorted_bank


# Make a new dataframe only using necessary columns / Change the necessary data to int type Customer Reference/Account number
final_bank = sorted_bank[['Account Number', 'Post Date', 'Transaction Description', 'Amount','Transaction Detail', 'Customer Reference']]

final_bank['Customer Reference'] = final_bank['Customer Reference'].fillna(0)
final_bank['Customer Reference'] = final_bank['Customer Reference'].astype(int)

final_bank['Account Number'] = final_bank['Account Number'].astype(str)
final_bank['Account Number'] = final_bank['Account Number'].str[-4:]
final_bank['Account Number'] = final_bank['Account Number'].astype(int)

final_bank.rename(columns={'Customer Reference': 'Check number'}, inplace=True)

final_bank
```

```python
check.head(15)
```

```python
# Organize the checkbook information (making columns and extract necessary columns)

check2 = check.iloc[8:].reset_index(drop=True)
check2 = check2.rename(columns=check2.iloc[0]).drop(check2.index[0]).reset_index(drop=True)
check2 = check2.dropna(subset=['Checkbook ID'])
check2 = check2.dropna(subset=['Amount'])
check2 = check2[['Check Number',	'Vendor ID',	'Vendor Check Name',	'Check Date',	'Checkbook ID',	'Audit Trail Code',	'Amount']]


# Getting Entity number from Checkbook ID column / change it to int type
check2['3Digits'] = check2['Checkbook ID'].str[:3]
check2 = check2.dropna(subset=['3Digits'])
check2['3Digits'] = pd.to_numeric(check2['3Digits'], errors='coerce')
check2['Check Number'] = pd.to_numeric(check2['Check Number'], errors='coerce')
check2['Check Number'] = check2['Check Number'].fillna(0).astype(int)

check2
```

```python
# Add Entity account number to Checkbook data

check3 = pd.merge(check2, bankaccount_subset[['Entity code', 'Account Number']], left_on='3Digits', right_on='Entity code', how='left')
check3
```

```python
# rows_with_nan = check3[pd.isna(check3['Account Number'])]
# rows_with_nan
```

```python
# Extract last 4 digist from Account Number / Change it to int type / Change the Check number column to int type

check3 = check3.dropna(subset=['Account Number'])
check3['Account Number'] = check3['Account Number'].str[1:5].astype(int)
check3['Check Number'] = check3['Check Number'].astype(int)
check3
```

```python
check4 = pd.merge(check3, final_bank[['Account Number', 'Check number','Post Date', 'Amount']], left_on=['Account Number', 'Check Number'], right_on=['Account Number', 'Check number'], how='left')
check4
```
